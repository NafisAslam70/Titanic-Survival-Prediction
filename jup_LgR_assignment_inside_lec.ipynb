```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
dd=pd.read_csv('LgR_train.csv')
dd.shape
```

```python
dd
```

```python
dd.drop('Cabin',axis=1, inplace=True)
dd.drop('Name',axis=1, inplace=True)
dd.Age.fillna(dd.Age.mean(),inplace=True)
```

```python
dd
```

```python
dd.dropna(inplace=True)
dd.isna().sum()

```

```python
dd.loc[dd.Sex == 'male', 'Sex'] = 1
dd.loc[dd.Sex == 'female', 'Sex'] = 0
dd['Age'] = dd['Age'].astype(int)
dd['Fare'] = dd['Fare'].astype(int)

dd
```

```python
dd.loc[dd.Embarked == 'S', 'Embarked'] = 0
dd.loc[dd.Embarked == 'Q', 'Embarked'] = 1
dd.loc[dd.Embarked == 'C', 'Embarked'] = 2
dd
```

```python
df=dd.copy()
df
```

```python
y=df.Survived
x=df.drop('Survived',axis=1)
```

```python
from sklearn import model_selection
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2, random_state=42)
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
machine=LogisticRegression(max_iter=1000)
```

```python
machine.fit(x_train,y_train)
```

```python
machine.predict(x_test)
```

```python
machine.score(x_test,y_test)
```
